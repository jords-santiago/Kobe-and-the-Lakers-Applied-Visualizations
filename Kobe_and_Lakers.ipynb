This is the project that I submitted in one of the courses of the Applied Data Science with Python Specialization in Coursera.

This visualization is with regards to the region Los Angeles, California, United States with category Sports and Athletics.

This project is trying to determine:

How has the win percentage of the Los Angeles Lakers change along with the performance (true shooting percentage) of Kobe Bryant throughout his 20 years with the Lakers?

The datasets are found in:

For Lakers record: Team_Records.csv from https://www.kaggle.com/boonpalipatana/nba-season-records-from-every-year

For Kobe Bryant's statistics: all_seasons.csv from https://www.kaggle.com/justinas/nba-players-data

```python
import pandas as pd
```

```python
lakers_df = pd.read_csv('Team_Records.csv')
kobe_df = pd.read_csv('all_seasons.csv')
```

```python
lakers_df = lakers_df[['Season', 'Team', 'W/L%']]
lakers_df['Team'] = lakers_df['Team'].str.replace("\*","")
lakers_df = lakers_df[lakers_df['Team'] == 'Los Angeles Lakers']
lakers_df.head(25)
```

```python
kobe_df = kobe_df[['player_name', 'ts_pct', 'season']]
kobe_df = kobe_df[kobe_df['player_name'] == 'Kobe Bryant']
kobe_df.columns = ['Player_Name', 'TS_PCT', 'Season']
kobe_df.head(20)
```

```python
kobe_lakers_df = pd.merge(kobe_df, lakers_df, how='inner', on = ['Season'])
kobe_lakers_df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,7))
plt.plot(np.arange(len(kobe_lakers_df)), kobe_lakers_df['W/L%'], linestyle='--', marker='o', label="Lakers' Record")
plt.plot(np.arange(len(kobe_lakers_df)), kobe_lakers_df['TS_PCT'], linestyle='-', marker='x', label="Kobe's True Shooting")

plt.xlabel('Season',size=14)
plt.ylabel('Percentage',size=14)
plt.xticks(np.arange(len(kobe_lakers_df)), kobe_lakers_df['Season'], rotation=45)

plt.title('The Los Angeles Lakers with Kobe Bryant (1996-2016)',size=14)
plt.legend(loc=0)

plt.show()
```

To answer the research question, both the Los Angeles Lakers' team record and Kobe Bryant's true shooting percentage were extracted from different datasets. The time period of the data is from 1996-2016 which covers Kobe Bryant's entire career with the Lakers. With this 20 year time period, this would suffice to gather relevant insights. True shooting percentage is used as it considers all levels of scoring for measuring a player's performance.

The plot shows the Lakers' winning percentage along with Kobe Bryant's true shooting percentage. These two can easily be compared as both are shown in terms of percentage. Based on the plot generated, Kobe's true shooting percentage is very much consistent throughout his career and only dipped in his last few seasons. This is consistent with the fact that he has sustained injuries due to wear and tear of his body leading to his retirement. On the other hand, the Lakers' team record has been above 500 (more wins than losses) for the most part but has dipped also during Kobe's last few seasons. With this, we can say that Kobe's performance is one of the factors of the Lakers' team success.
